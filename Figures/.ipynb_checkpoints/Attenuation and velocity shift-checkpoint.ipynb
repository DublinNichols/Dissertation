```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np

# Define the functions F1 and F2
def v(x):
    return 1/(1+(1/x)**2)

def G(x):
    return 1/x/(1+(1/x)**2)

# Generate x values (logarithmic scale)
x_values = np.linspace(0.01, 100, 10000)  # Adjust the range as needed

# Calculate corresponding y values for F1 and F2
y_values_F1 = v(x_values)
y_values_F2 = G(x_values)

# Create the plot
# Create the plot with dual y-axes
fig, ax1 = plt.subplots(figsize=(5,3))

matplotlib.rcParams['xtick.minor.size'] = 0
matplotlib.rcParams['xtick.minor.width'] = 0

# Left y-axis for F1
ax1.plot(x_values, y_values_F2, label='v', color='black')
ax1.set_xlabel(r"$\omega_c/\omega}$")
ax1.set_ylabel(r"$\Gamma \: \lambda_{SAW}/K^2} \; \mathrm{(m^{-1})}$", color='black')
ax1.tick_params(axis='y', labelcolor='black')

# Right y-axis for F2
ax2 = ax1.twinx()
ax2.plot(x_values, y_values_F1, label='F2(x)', color='red')
ax2.set_ylabel(r"$\Delta v/(v_0 K^2)$", color='red')
ax2.tick_params(axis='y', labelcolor='red')
plt.xscale('log')

ax1.set_xticks([0.01, 0.1, 1, 10, 100])
ax1.set_xticklabels(["0.01", "0.1", "1", "10", "100"])





plt.savefig(r"C:\Users\Dublin\Box\Minot Group Documents\Dublin\Documents\Dissertation\Figures\gamma_and_v.pdf", format="pdf", bbox_inches = "tight")
plt.show()
```
