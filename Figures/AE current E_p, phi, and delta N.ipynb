```python
import numpy as np
import matplotlib.pyplot as plt



fig, ax = plt.subplots(figsize = (6.7,8), dpi = 300, nrows = 3)



def E_p(x, l, p):
    return np.cos(l*x + p) + 1j*np.sin(l*x + p)


def E_eff(x, l, p):
    return E_p(x,l,0)/(1+ 1j*p)/np.absolute(E_p(x,l,0)/(1+ 1j*p))

def phi(x, l):
    return np.sin(l*x)



#(np.cos(l*x + p) + 1j*np.sin(l*x+p)))

# def E_p(x, l):
#     return np.cos(l*x) + 1j*np.sin(l*x)


# def E_eff(x, l, p):
#     return (E_p(x,l)/(1+ 1j*p))*np.conjugate(E_p(x,l)/(1+ 1j*p))

# def phi(x, l):
#     return np.sin(l*x)


l = 2*np.pi

x = np.arange(-1,1,0.01)

# pshift = [np.arctan(0.001), np.arctan(1), np.arctan(1000000000)]

pshift = [0, 1]

names = [r"$\sigma_{2D} \ll \sigma_M$", r"$\sigma_{2D} = \sigma_M$", r"$\sigma_{2D} \gg \sigma_M$"]

for axnum, p in enumerate(pshift):
    ax2 = ax[axnum].twinx()  # Create a second y-axis
    
    ax2.yaxis.tick_right()  # Make the second y-axis visible only on the right
    
    ax2.yaxis.set_tick_params(color='blue')  # Change the color of the ticks
    
    ax2.spines['right'].set_color('blue')  # Change the color of the axis

    ax[axnum].plot(x,E_p(x, l,0), color = "black", label = r"$E_p$")
    
    ax[axnum].plot(x,E_eff(x, l, p), color = "blue", ls = "--", label = r"$\Delta_n$")
    
    ax[axnum].set_title(names[axnum])

    ax[axnum].set_yticks([-1,0,1])
    
    ax[axnum].set_xticklabels([])
    
    ax[axnum].tick_params(direction = "in", top = True)

    ax2.set_ylim((-0.125, 1.125))
    ax2.set_yticks([0,0.5,1])
    
    
    ax2.tick_params(direction = "in")
    
ax[2].plot(x,E_p(x, l,0), color = "black", label = r"$E_p$")
ax[2].plot(x,x*0, color = "blue", ls = "--", label = r"$\Delta_n$")
    
plt.savefig("piezoelectric field and delta n.svg")
# for axnum, p in enumerate(pshift):
    
 
# #     ax[axnum].plot(x,phi(x, l), color = "blue", label = r"$\phi$")
    
#     ax[axnum].plot(x,E_p(x, l,0), color = "black", label = r"$E_p$")

#     #r"$\mathrm{\Delta}n$")
        
#     ax[axnum].plot(x,E_eff(x, l, p), color = "blue", ls = "--", label = r"$\Delta n$")

    
#     ax[axnum].set_title(names[axnum])
# #     ax[axnum].legend(loc = "lower left")

#     ax[axnum].set_yticklabels([])
#     ax[axnum].set_xticklabels([])
    
#     ax[axnum].tick_params(direction ="in", top = True, right = True)
```
