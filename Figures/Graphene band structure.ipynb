```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the function to be plotted
def f(x, y):
    return np.sin(np.sqrt(x**2 + y**2))

# Generate x, y, z values
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = f(x, y)

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(x, y, z, color='b', alpha=1)

# Plot a line on the surface where y = 1
x_line = np.linspace(-5, 5, 100)
y_line = np.ones_like(x_line)
z_line = f(x_line, y_line)
ax.plot(x_line, y_line, z_line, color='red', linewidth=2)

# Adjust the view angle
ax.view_init(elev=20, azim=-35)

plt.savefig("line on surface test.svg")
```
