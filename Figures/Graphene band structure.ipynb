```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D, art3d
import numpy as np
from matplotlib.patches import Rectangle

npts = 5000
# Create data
x = np.linspace(-3, 3, npts)
y = np.linspace(-3, 3, npts)
x, y = np.meshgrid(x, y)


a = 0.142

def E(kx, ky):
    E = (3 + 2*np.cos(3**(1/2)*kx)+4*np.cos((3/2)*ky)*np.cos(3**(1/2)*kx/2))**(1/2)
    return E

# Create a figure for the 3D plot
fig = plt.figure(figsize=(12, 6), dpi = 300)

# Add a 3D subplot
ax1 = fig.add_subplot(121, projection='3d')
ax1.plot_surface(x, y, E(x,y), color = "grey")
ax1.plot_surface(x, y, -E(x,y), color = "grey")
ax1.set_xlabel(r"$k_x a$")
ax1.set_ylabel(r"$k_y a$")
ax1.set_zlabel(r"$E/t$")
ax1.view_init(elev = 10, azim = -60)


xkpt = 2*np.pi/3
ykpt = 2*np.pi/(3*3**(1/2))

zoomn = 1
# Zoom in on a certain region in a 2D plot
zoom_x = [xkpt - zoomn, xkpt + zoomn]
zoom_y = [ykpt - zoomn, ykpt + zoomn]
zoom_z = [-0.5, 0.5]

# Add a 2D subplot for the zoomed region
ax2 = fig.add_subplot(122, projection = "3d")

zoomn2 = 2
xstart, xstop = xkpt - zoomn2, xkpt + zoomn2
ystart, ystop = ykpt - zoomn2, ykpt + zoomn2
x2 = np.arange(xstart, xstop, 0.01)
y2 = np.arange(ystart, ystop, 0.01)
x2, y2 = np.meshgrid(x2, y2)

ax2.plot_surface(x2, y2, E(x2,y2), color = "grey")
ax2.plot_surface(x2, y2, -E(x2,y2), color = "grey")
ax2.set_xlim(zoom_x)
ax2.set_ylim(zoom_y)
ax2.set_zlim(zoom_z)
ax2.view_init(elev = 0, azim = 0)
ax2.axis("off")
ax2.grid(False)

# Indicate the zoomed region on the 3D plot
# Create a rectangle patch
# rect = Rectangle((zoom_x.min(), zoom_y.min()), zoom_x.max()-zoom_x.min(), zoom_y.max()-zoom_y.min(), linewidth=1, edgecolor='r', facecolor='none')

# Add the patch to the Axes
# ax1.add_patch(rect)

# Convert (x, y) to 3D space
# art3d.pathpatch_2d_to_3d(rect, z=zoom_z.min(), zdir="z")

plt.savefig("Graphene band structure 3D plot.png")
plt.show()
```
